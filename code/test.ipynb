```python
import pandas as pd
import geopandas as gpd
import pyarrow.feather as feather
from shapely import wkt

def read_feather(file_path):
    # 读取feather文件
    df = pd.read_feather(file_path)

    # 将WKT格式的几何列(geometry)转换回地理信息
    df['geometry'] = df['geometry'].apply(lambda x: wkt.loads(x))

    # 将pandas的DataFrame对象转换为geopandas的GeoDataFrame对象
    gdf = gpd.GeoDataFrame(df, geometry='geometry')

    # 设置坐标参考系统（CRS）
    # 请根据实际情况替换为您的坐标参考系统
    gdf.set_crs("EPSG:28413", inplace=True)

    return gdf

# 将新方法添加到geopandas模块
gpd.read_feather = read_feather

x = gpd.read_feather('CL/lu.feather')
x
```

```python
import csv
 
# 假设我们要提取的列是第二列，索引为1
column_to_check = 3
value_to_find = 'Power'
 
input_file = 'China.csv'
output_file = 'output.csv'
 
with open(input_file, 'r', newline='') as csv_in_file, open(output_file, 'w', newline='') as csv_out_file:
    csv_reader = csv.reader(csv_in_file)
    csv_writer = csv.writer(csv_out_file)
    
    # 读取标题行
    headers = next(csv_reader)
    csv_writer.writerow(headers)
    
    # 筛选出特定列值为1的记录
    for row in csv_reader:
        if row[column_to_check] == str(value_to_find):
            csv_writer.writerow(row)
```

```python
import pandas as pd
import glob
import os

# 定义Excel文件夹的路径
folder_path = "prob1"

# 获取文件夹内所有Excel文件的路径
excel_files = glob.glob(os.path.join(folder_path, "*.xlsx"))

# 初始化一个空的DataFrame来存储拼接结果
combined_df = pd.DataFrame()

# 循环读取每个Excel文件并拼接
for file in excel_files:
    df = pd.read_excel(file, usecols=['pointid', 'PROB_GENG ', 'PROB_LIN'])  # 只读取指定的三列
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# 将拼接后的DataFrame保存为一个新的Excel文件
output_path = os.path.join(folder_path, "combined.xlsx")
combined_df.to_excel(output_path, index=False)

print(f"All files have been successfully combined and saved to {output_path}")
```

```python
import os
import pandas as pd

# 指定包含所有CSV文件的文件夹路径
folder_path = 'Ws1'

# 获取文件夹内所有CSV文件的文件名
all_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# 初始化一个空的DataFrame来存储所有CSV文件的内容
combined_data = pd.DataFrame()

# 逐个读取CSV文件并拼接数据
for file in all_files:
    file_path = os.path.join(folder_path, file)
    data = pd.read_csv(file_path)
    combined_data = pd.concat([combined_data, data], ignore_index=True)

# 打印拼接后的数据
#print(combined_data)

# 如果需要保存拼接后的数据到新的CSV文件中，可以使用下面的代码
combined_data.to_csv('Ws1/4.csv', index=False)
```
